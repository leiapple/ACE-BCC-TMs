```python
# Import the function
from dislocation_analysis import calculate_dislocation_properties

# Calculator setup 
from pyace import PyACECalculator
```

```python


# Create calculator
calc_Mo = PyACECalculator('../ACE2025_potential/Mo/output_potential.yaml')
calc_Nb = PyACECalculator('../ACE2025_potential/Nb/output_potential.yaml')
calc_Ta = PyACECalculator('../ACE2025_potential/Ta/output_potential.yaml')
calc_V =  PyACECalculator('../ACE2025_potential/V/output_potential.yaml')
calc_W =  PyACECalculator('../ACE2025_potential/W/output_potential.yaml')

# Define parameters
elements = [ 'W', 'Mo', 'V', 'Ta', 'Nb']
a0_list = [ 3.185, 3.163, 2.99, 3.319, 3.307]
calculators = [calc_W, calc_Mo, calc_V, calc_Ta, calc_Nb]

# Run the analysis
results = calculate_dislocation_properties(
    elements=elements,
    a0_list=a0_list,
    calculators=calculators,
    fmax=0.005,
    optimizer='BFGS'
)

# Display results
print(results)

# Save to CSV
results.to_csv('dislocation_results.csv', index=False)
```
